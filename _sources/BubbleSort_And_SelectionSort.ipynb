**1. BubbleSort**

Buat modifikasi algoritma bubble sort. Jika pada algoritma bubble sort, pointer bergerak
dari kiri ke kanan (menuju index terakhir), sedemikian hingga data terbesar berada pada
index terakhir. Modifikasi yang dilakukan adalah, setelah data terbesar terletak pada in-
dex terakhir. Pointer bergerak lagi dengan arah sebaliknya, yaitu dari kanan ke kiri untuk
membawa data terkecil ke index paling awal (bidirectional bubble sort)

```python
def bubbleSort(listData):
    c=0
    print('Data yang akan diurutkan : ', listData)
    for a in range(len(listData)-1,0,-1):
        c+=1 
        print ('Iterasi ke-', c,':')
        for b in range(len(listData)-1, (len(listData)-a)-1, -1):
            if listData[b]<listData[b-1]:
                placement=listData[b]
                listData[b]=listData[b-1]
                listData[b-1]=placement
            print(listData)
    print('Data urut-',listData)
b=[12,1,15,25,8,6]

bubbleSort(b)
```

Modifikasi code selection sort tersebut, agar iterasi dalam melakukan dua buah pencarian,
yaitu nilai minimal dan maksimal, jika sudah ditemukan, maka letakkan nilai minimal pada
indeks-indeks awal, dan nilai maksimal pada indeks-indeks akhir
