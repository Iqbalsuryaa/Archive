6. Tuliskan suatu program Python untuk menerima nama file dari pengguna dan cetak ekstensi dari 
file tersebut. 
Contoh nama file: abc.java 
Output : java

```python
#6. Tuliskan suatu program Python untuk menerima nama file dari pengguna dan cetak ekstensi dari 
#file tersebut. 
#Contoh nama file: abc.java 
#Output : java

namaFile = input("masukkan nama file: ")
n = len(namaFile)-1
ext = ""
result = ""

while namaFile[n] != ".":
	ext += namaFile[n]
	n -= 1

for i in range(len(ext)-1,-1,-1):
	result += ext[i]

print("nama file", namaFile)	
print("extensinya adalah", result)
```

8. Tuliskan suatu program yang menampilkan luas dan keliling dari suatu lingkaran yang mepunyai 
radius (jari-jari) 5.5 menggunakan rumus berikut (Catatan: π = 3.14159): 
luas = 𝑟𝑎𝑑𝑖𝑢𝑠 × 𝑟𝑎𝑑𝑖𝑢𝑠 × 𝜋
keliling = 2 × 𝑟𝑎𝑑𝑖𝑢𝑠 × 𝜋

```python
#8. Tuliskan suatu program yang menampilkan luas dan keliling dari suatu lingkaran yang mepunyai 
#radius (jari-jari) 5.5 menggunakan rumus berikut (Catatan: π = 3.14159): 
#luas = 𝑟𝑎𝑑𝑖𝑢𝑠 × 𝑟𝑎𝑑𝑖𝑢𝑠 × 𝜋
#keliling = 2 × 𝑟𝑎𝑑𝑖𝑢𝑠 × 𝜋

r = 5.5
pi = 3.14159

luas = r*r*pi
keliling = 2*r*pi
print('luas lingkaran =',int(luas))
print('keliling lingkaran =',int(keliling))
```
