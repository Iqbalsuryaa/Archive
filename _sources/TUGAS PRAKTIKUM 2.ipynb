2. (Menjumlahkan digit-digit dalam bilangan bulat)
   Tuliskan suatu program yang membaca bilangan integer antara 0 dan      1000   dan
 kemudian menambahkan semua digit di dalam integer tersebut. Sebagai contoh,
jika nilai integernya adalah 932, maka jumlah semua digitnya adalah 14. 

   (Hint:Gunakan operator % untuk mengekstrak digit, dan gunakan operator // untuk menghapus digit-digit yang diekstrak. Misalnya, 932 % 10 = 2 dan 932 // 10 = 93.)

   Di bawah ini adalah contoh tampilan saat program berjalan:
Masukkan angka antara 0 dan 1000: 999
Jumlah digitnya adalah 27

```python
#(Menjumlahkan digit-digit dalam bilangan bulat)

varA=int(input("masukkan angka antara 0 dan 1000 : "))

jumlah=0

for digit in str(varA):
    jumlah=jumlah+int(digit)

print("jumlah digitnya adalah",jumlah)
```

3. (Fisika: mencari panjang runway (landasan pacu))

dapat menghitung panjang landasan pacu minimal yang diperlukan agar suatu
pesawat udara dapat take off secara aman dengan menggunakan formula
berikut:

 panjang = v 2 / 2a

 Tuliskan suatu program yang meminta pengguna untuk memasukkan v dalam
 satuan meters/second (m/s) dan akselerasi a dalam meters/second kuadrat
 (m/s2), dan kemudian menampilkan panjang minimum dari landasan tersebut.
 
 Berikut ini adalah contoh tampilan saat program berjalan:
 Masukkan kecepatan dan akselerasi: 60, 3.5 <enter>
 Panjang minimal landasan pacu untuk pesawat ini adalah 514.286 meter.

5. (Geometri: Luas suatu segitiga) Tuliskan suatu program yang meminta pengguna
memasukkan tiga titik koordinat dari suatu segitiga (x1, y1), (x2, y2), dan (x3, y3) dan
kemudian tampilkan luas dari segitiga tersebut.

Rumus untuk menghitung luas (area) segitiga adalah

Berikut ini adalah hasil saat program dijalankan:
Masukkan tiga titik dari suatu segitiga: 1.5, -3.4, 4.6, 5, 9.5, -3.4
Luas segitiga ini adalah 33.6

Catatan: Anda dapat menggunakan rumus lain untuk menghitung luas segitiga ini, di
antaranya:


```python
#Geometri: Luas suatu segitiga

x1,y1,x2,y2,x3,y3 = eval(input("Masukkan tiga titik dari suatu segitiga: "))

luas_segitiga = abs((0.5)*(x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2)))
print("Luas segitiga ini adalah %.1f" % luas_segitiga)
```

```python
#Geometri: Luas suatu segitiga

x1, y1, x2, y2, x3, y3 = eval(input("Masukkan tiga titik dari suatu segitiga: "))
side1 = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
side2 = ((x3 - x1) ** 2 + (y3 - y1) ** 2) ** 0.5
side3 = ((x2 - x3) ** 2 + (y2 - y3) ** 2) ** 0.5
s = (side1 + side2 + side3) / 2
area = (s * (s - side1) * (s - side2) * (s - side3)) ** 0.5
print("Luas segitiga ini adalah %.1f" % area)
```

6. (Aplikasi keuangan: nilai majemuk) Katakanlah anda menyimpan $100 setiap bulan ke

itu, bunga bulanannya menjadi 0.00417. Setelah bulan pertama, nilai uang anda akan
menjadi: 100 * (1 + 0.00417) = 100.417
Setelah bulan kedua, dimana anda menambahkan tabungan bulanan $100, tabungan anda menjadi (100 + 100.417) * (1 + 0.00417) = 201.252
Setelah bulan ketiga, tabungan anda menjadi (100 + 201.252) * (1 + 0.00417) = 302.507 Dan seterusnya.
Buatlah suatu program yang meminta pengguna memasukkan besarnya tabungan
bulanannya dan tampilkan total tabungan yang dimiliki setelah enam bulan.
Berikut ini adalah contoh keluaran dari program tersebut:
Masukkan nilai tabungan bulanan: 100 <enter>
Setelah enam bulan, tabungan anda bernilai 608.81

```python
#Aplikasi keuangan: nilai majemuk

bunga = 0.00417

tabungan = float(input("Masukkan nilai tabungan bulanan: "))
bulan1 = tabungan*(1+ bunga )
bulan2 = ( tabungan + bulan1 )*(1+ bunga )
bulan3 = ( tabungan + bulan2 )*(1+ bunga )
bulan4 = ( tabungan + bulan3 )*(1+ bunga )
bulan5 = ( tabungan + bulan4 )*(1+ bunga )
bulan6 = ( tabungan + bulan5 )*(1+ bunga )

print("Setelah enam bulan, tabungan anda bernilai = ", bulan6 )
```
