Tuliskan suatu program yang meminta pengguna untuk memasukkan a, b, c, d, e, dan f dan 
kemudian tampilkan hasilnya. Jika ad â€“ bc bernilai 0, laporkan bahwa Persamaan tidak mempunyai 
solusi. 
Berikut ini adalah contoh keluaran dari program tersebut: 
Masukkan a, b, c, d, e, f: 9.0, 4.0, 3.0, -5.0, -6.0, -21.0 x = -2.0 dan y = 3.0 
Masukkan a, b, c, d, e, f: 1.0, 2.0, 2.0, 4.0, 4.0, 5.0 Persamaan tidak mempunyai solusi.

```python
#Code nomor 2

a,b,c,d,e,f=eval(input("Masukkan a, b, c, d, e, f: "))

if a*d-b*c==0:
    print('Persamaan tidak mempunyai solusi')
else:
    x=(e*d-b*f)/(a*d-b*c)
    y=(a*f-e*c)/(a*d-b*c)
    print("x =",x,"and y =",y)
```
