1. (Menampilkan piramida)

Tuliskan suatu program yang meminta pengguna memasukkan suatu bilangan bulat
(integer) dari 1 sampai 15 dan tampilkan suatu piramida, sebagaimana diperlihatkan
pada contoh eksekusi di bawah ini:

Masukkan jumlah baris: 7

```python
n = int(input('Masukkan jumlah baris: '))  

for i in range(n):     
    spaces = ' ' * (2*(n-1-i))     
    print(spaces, end='')
          
    for j in range(i+1, 0, -1):         
        print('%2d' % j, end='')

    for j in range(2, i+2, 1):         
        print('%2d' % j, end='')
             
    print()
```

```python
def futureInvestmentValue(investmentAmount, monthlyInterestRate, years):     
    return investmentAmount * (1+monthlyInterestRate)**(years*12)   
    
def main():     
    amount = float(input('Jumlah yang diinvestasikan: '))     
    interest = float(input('Tingkat Bunga Tahunan: '))      
    print('\nTahun\tNilai Masa Depan')     
    for i in range(1, 31):         
        print(str(i) + '\t' + '%.2f' % futureInvestmentValue(amount, interest/1200, i))        
main() 
```

7. (Konversi antara kaki dan meter)
Buatlah suatu modul yang mengandung dua fungsi berikut:



```python
foot = 1
meter = 20

def footToMeter(foot):
    return foot *0.305

def meterToFoot(meter):
    return round((meter/0.305),3)

print('Kaki\tMeter\t|\tMeter\tKaki')
for foot,meter in zip(range(1, 11), range(20, 75, 6)):
    print('%s\t%s\t|\t%s\t%s' % (foot, footToMeter(foot), meter, meterToFoot(meter)))
```

9. (Estimasi )
Estimasi dapat dihitung menggunakan rumus deret berikut:

Tulislah suatu fungsi yang mengembalikan m(i) dari nilai i tertentu dan buatlah suatu
program test yang menampilkan tabel berikut:
i m(i)
1 4.0000
101 3.1515
201 3.1466
301 3.1449
401 3.1441
501 3.1436
601 3.1433
701 3.1430
801 3.1428
901 3.1427

```python
def calPi(i): 
    mi = 0 
    for j in range(1,i+1):
        mi += 4*((-1)**(j+1))/(2*j-1) 
    return mi

print('i\tm(i)')
for i in range(1,1000,100):
    mi = calPi(i)
    print("{0:d}\t{1:.4f}".format(i, mi))
```
