(Geografi: Memperkirakan luas)
Carilah lokasi GPS dari kota Atlanta, Georgia; Orlando, Florida; Savannah, Georgia;
dan Charlotte, North Carolina dari www.gps-data-team.com/map/ dan hitung
perkiraan luas yang dilingkupi oleh keempat kota ini.
Catatan:
• Gunakan rumus pada masalah 5 di atas untuk menghitung jarak antara 2 kota
• Bagi polygon menjadi dua segitiga dan gunakan rumus berikut untuk
menghitung luas segitiga tersebut

```python
# nomor 2
#codee

import math

Atlanta_x, Atlanta_y = 33.753746, -84.386330
Savannah_x, Savannah_y = 28.5383355, -81.3792364999
Orlando_x, Orlando_y = 32.0835407, -81.09983419999998
Charlotte_x, Charlotte_y = 35.2270869, -80.84312669999997


avg = 6371.01

#jarak antara Atlanta dan Savannah di permukaan bumi
AS = avg * math.acos (
    math.sin(math.radians(Atlanta_x))
    * math.sin(math.radians(Savannah_x))
    + math.cos(math.radians(Atlanta_x))
    * math.cos(math.radians(Savannah_x))
    * math.cos(math.radians(Atlanta_y - Savannah_y)))

#jarak antara Atlanta dan Charlotte di permukaan bumi
AC = avg * math.acos (
    math.sin(math.radians(Atlanta_x))
    * math.sin(math.radians(Charlotte_x))
    + math.cos(math.radians(Atlanta_x))
    * math.cos(math.radians(Charlotte_x))
    * math.cos(math.radians(Atlanta_y - Charlotte_y)))

#jarak antara Savannah dan Charlotte di bumi
SC = avg * math.acos (
    math.sin(math.radians(Savannah_x))
    * math.sin(math.radians(Charlotte_x))
    + math.cos(math.radians(Savannah_x))
    * math.cos(math.radians(Charlotte_x))
    * math.cos(math.radians(Savannah_y - Charlotte_y)))

#jarak antara Atlanta dan Orlando di permukaan bumi
AO = avg* math.acos (
    math.sin(math.radians(Atlanta_x))
    * math.sin(math.radians(Orlando_x))
    + math.cos(math.radians(Atlanta_x))
    * math.cos(math.radians(Orlando_x))
    * math.cos(math.radians(Atlanta_y - Orlando_y)))

#jarak antara Savannah dan Orlando di permukaan bumi
SO = avg * math.acos (
      math.sin(math.radians(Savannah_x))
    * math.sin(math.radians(Orlando_x))
    + math.cos(math.radians(Savannah_x))
    * math.cos(math.radians(Orlando_x))
    * math.cos (math.radians(Savannah_y - Orlando_y)))
#luas segitiga antara Atlanta, Savannah dan Charlotte
s = (AS + AC + SC) / 2 
area1 = (s * (s - AS) * (s - AC) * (s - SC)) ** 0.5
print('luas segitiga 1 :', area1)

#luas segitiga antara Atlanta, Savannah dan Orlando
s = (AS + AO + SO) / 2
area2 = ( s * (s - AS) * (s - AO) * (s - SO)) ** 0.5
print('luas segitia 2 :', area2)

area = area1 + area2

#display area poligon yang dibuat oleh 4 kota
print("Perkiraan luas daerah yang dilingkupi oleh 4 kota adalah : " + str(area), "sq.km")
```

(Geometri: luas polygon beraturan)
Poligon beraturan adalah poligon bersisi n di mana semua sisinya sama panjang dan
semua sudutnya memiliki derajat yang sama (yaitu, poligonnya sama sisi dan sama
sudut). Rumus untuk menghitung luas poligon beraturan adalah

Di sini, s adalah panjang sisi.

Tulis program yang meminta pengguna untuk memasukkan jumlah sisi dan panjang sisi
tersebut dari suatu poligon beraturan dan tampilkan luasnya.
Berikut ini adalah contoh eksekusinya:
Masukkan jumlah sisi: 5
Masukkan panjang sisi: 6.5
Luas poligon adalah 73.69017017488385

```python
# nomor 3

import math
n = eval(input("Masukkan jumlah sisi: "))
s = eval(input("Masukkan panjang sisi: "))
area = (n*(s**2))/(4*math.tan(math.pi/n))
print("Luas poligon adalah ",area)
```
