Kode program SubtractionQuizLoop.py membangkitkan lima pertanyaan pengurangan acak.
Lakukan revisi terhadap program itu untuk membangkitkan 10 pertanyaan penambahan acak
terhadap dua bilangan bulat antara 1 dan 15. Tampilkan jumlah yang betul dan waktu testnya.

```python
# nomor 1

import time
import random
correctCount = 0  #Count the number of correct answers
count = 0 #Count the number of questions
startTime = time.time()*1000.0
while (count < 10):
    number1 = random.randint(1,16)
    number2 = random.randint(1,16)
    #Prompt the student to answer "what is number1 â€“ number2?"
    print("Berapakah  {}  -  {} ? ".format(number1,number2))
    answer = int(input())
    #Grade the answer and display the result
    if (number1 - number2 == answer):
        print( "Jawaban mu benar!")
        correctCount=correctCount+1
    else:
        print("jawabanmu salah.\n {} - {}  yang benar adalah : ".format(number1,number2), (number1 - number2))
    #Increase the count by incremeting variable called "count"
    count=count+1
print("jawaban yang benar : ",correctCount)
endTime = time.time()*1000.0
print("Waktumu adalah  {} detik".format((endTime - startTime) / 1000))
```

(Menggunakan fungsi trigonometris)
Cetak tabel berikut untuk menampilkan nilai sin dan cos dari derajat mulai 0 sampai 360
dengan kenaikan sebesar 10 derajat. Bulatkan nilai untuk tetap empat digit setelah titik
desimal.

```python
# nomor 2

import math

derajat = 0
sin = 0
cos = 1.0

print( '%10s %10s %10s\n' %('Derajat', 'Sin', 'Cos') )

while derajat<=360:
    sin=math.sin(math.radians(derajat))
    cos=math.cos(math.radians(derajat)) 
    print( '%10.0f %10.4f %10.4f' %(derajat, sin, cos))
    derajat += 10
```

(Aplikasi keuangan: menghitung SPP masa depan)

biaya sekolah 4 selama 4 tahun terhitung sejak 10 tahun mendatang.

(Mencari bilangan yang habis dibagi 5 atau 6, tetapi tidak keduanya)
Tuliskan suatu program yang menampilkan 10 bilangan per baris, semua bilangan dari 100
sampai 200 yang dapat dibagi dengan 5 atau 6, tetapi tidak keduanya. Bilangan-bilangan
tersebut dipisahkan dengan pasti satu spasi.

```python
# nomor 6

bilangan = 0
hasil = ''

for x in range(100,201):
    if x % 30 == 0:
        continue
    if x % 5 == 0:
        hasil = hasil + str(x) + ' '
        bilangan = bilangan + 1
    elif x % 6 == 0:
        hasil = hasil + str(x) + ' '
        bilangan = bilangan + 1
    
    if bilangan == 10:
        print(hasil)
        bilangan = 0
        hasil = ''
    elif x == 200:
        print(hasil)
```
